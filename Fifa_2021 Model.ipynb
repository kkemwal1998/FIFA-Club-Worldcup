Based on the above, we will be drawing a relationship between the player's performance in the Fifa tournament and their net worth. For this reason, we will converting the data type of player's net worth into integral format in order to avoid any inconsistencies. 


# Data-Cleaning:

```python
FIFA
```

```python
# Removing the '€' in dataframe.
FIFA['Value'].replace('€','',regex=True,inplace=True)
FIFA
```

```python
# Converting the data type of a dataframe column Value.
def CONV_TYPE(value):
    if 'K' in value:
        k = value.replace('K',' ')
        A = pd.to_numeric(k)
        V1 = A*1000
        return float(V1)
    
    elif 'M' in value:
        m = value.replace('M',' ')
        B = pd.to_numeric(m)
        V2 = B*1000000
        return float(V2)
    
    else:
        return float(value)
    
```

```python
# Applyting the function
FIFA['Value'] = FIFA['Value'].apply(CONV_TYPE)
FIFA
```

```python
FIFA['Value'] = FIFA['Value'].astype(int)     # Changing the data-type of dataframe.
FIFA['Value']
```

```python
FIFA.info()
```

Therefore, we have convert the data type of player's net worth to int type. 

# Data-Modeling:

```python
FIFA.columns
```

```python

# Since we are drawing relationship between the net worth and thier performance in the match. 
# But before moving forward, we will draw a correlation between the varaibles in order to avoid overfitting of model.
Essential_variables = FIFA[['Value','Attacking','Crossing','Finishing','Heading Accuracy','Short Passing','Volleys','Dribbling','Curve','FK Accuracy','Long Passing','Ball Control', 'Movement','Acceleration', 'Sprint Speed', 'Agility', 'Reactions', 'Balance','Power', 'Shot Power', 'Jumping','Long Shots',
       'Mentality', 'Aggression', 'Interceptions', 'Positioning', 'Vision','Penalties', 'Composure', 'Defending', 'Marking', 'Standing Tackle','Sliding Tackle', 'Goalkeeping', 'GK Diving', 'GK Handling','GK Kicking', 'GK Positioning', 'GK Reflexes']]
Correlation = Essential_variables.corr()
Correlation
```

#After observing the relations between value and other variables,and stricly athering to the rule of thumb (https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3576830/). We will be moving with the variables that have scale levels between 0.3-1.0. Which are:-
1) Short Passing
2) Long Passing
3) Ball Control	
4) Reactions
5) Power
6) Shot Power
7) Mentality
8) Vision	
9) Composure


```python
# Independent/Explaratory variables.
X = FIFA[['Short Passing','Long Passing','Ball Control','Reactions','Power','Shot Power','Mentality','Vision','Composure']]
```

```python
# Dependent variable.
y = FIFA[['Value']]
```

Now we will be performing the train test split function to analyse the fit of equation. For this reason, we will be importing sklearn.


```python
from sklearn.model_selection import train_test_split
# We are preparing the training and testing dataset for further procedures that we have chosen for our reggression equations.
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=101)
# We have chosen test size as 0.3 that means that 30% of dataset shall be subject to testing.
```

```python
# Now we will be testing the dataset, in order to train and create a model.
from sklearn.linear_model import LinearRegression
```

```python
LM = LinearRegression() # Creating a model
```

```python
LM.fit(X_train,y_train)       # We will be fitting our model to train our dataset.
```

```python
LM.intercept_                    #Intercept value 
```

```python
LM.coef_                #Coefficient values of explaratory varaibles.
```

```python
X_train.columns
```

```python
# We will be creating a dataframe of the column values of x variables.
CDF = pd.DataFrame(LM.coef_ ,columns=X_train.columns)
CDF
```

```python
#################Predicted values of x variables.
```

```python
Prediction = LM.predict(X_test)
Prediction
```

```python
plt.figure(figsize=(13,15))
plt.scatter(y_test,Prediction)
plt.xlabel('Net worth')
plt.ylabel('Predicted Values')
plt.grid(True)
plt.show()
```

Insight:It can be concluded that there exists a weak positive relationship between the player's net worth and their perfomance in the match.
 

```python
plt.figure(figsize=(50,27))
sns.displot((y_test-Prediction))
plt.xlabel('Net Worth')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

Conclusion: Since our graph is normally distributed. Therefore, it means that our model that we have chosen wad suitable for our dataset.
